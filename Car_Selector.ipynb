```python
X_train = df[['Mileage', 'Engine_Power', 'Number_of_seats','Price']]
X_train
y_train = df.Car_Name
```

```python
#X_train = X_train.convert_objects(convert_numeric=True).dtypes
X_train.transform(pd.to_numeric, errors='coerce')
from sklearn import preprocessing
mini = df[['Mileage', 'Engine_Power', 'Number_of_seats','Price']].min()
maxi = df[['Mileage', 'Engine_Power', 'Number_of_seats','Price']].max()

```

```python
df[['Mileage', 'Engine_Power', 'Number_of_seats','Price']] = (df[['Mileage', 'Engine_Power', 'Number_of_seats','Price']] - mini / (maxi - mini))
df
mini
maxi
```

```python
import time
t0 = time.time()
from sklearn.ensemble import RandomForestClassifier
neigh = RandomForestClassifier()
# Predicted class
neigh.fit(X_train, y_train)
t1=time.time()
print(t1-t0)
```

```python
from sklearn.externals import joblib 
  
# Save the model as a pickle in a file 
joblib.dump(neigh, 'cars.pkl') 
```
